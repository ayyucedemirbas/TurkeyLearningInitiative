```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs, make_classification

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, plot_confusion_matrix, f1_score, matthews_corrcoef
from sklearn import tree
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv("assignment5_data.csv")
data.head()
```

## Q1.

```python
mis_data=data.copy()    
np.random.seed(3)    
indexes=np.random.choice(np.arange(len(data)),size=500, replace=False)   ##creates 500 missing value in x6
mis_data.loc[indexes,"x6"]=np.nan 

np.random.seed(5)
indexes=np.random.choice(np.arange(len(data)),size=500, replace=False)
mis_data.loc[indexes,"x9"]=np.nan

np.random.seed(11)
indexes=np.random.choice(np.arange(len(data)),size=500, replace=False)
mis_data.loc[indexes,"x11"]=np.nan

np.random.seed(9)
indexes=np.random.choice(np.arange(len(data)),size=500, replace=False)
mis_data.loc[indexes,"x13"]=np.nan
```

```python
x=mis_data.drop("target", axis=1)
y=mis_data["target"]
x_train, x_rest, y_train, y_rest = train_test_split(x,y, test_size=0.4, random_state=55)
x_valid, x_test, y_valid, y_test = train_test_split(x_rest,y_rest, test_size=0.5, random_state=38)
print(x_train.shape, x_valid.shape, x_test.shape, y_train.shape, y_valid.shape, y_test.shape )
```

```python
def report_missingness_percentages(data):
    for column in data.columns:
        if data[column].isnull().sum()>0:
            print(column, ">> missing nmuber:", data[column].isnull().sum())
            print(column, ">> missing percentage:", data[column].isnull().sum()/len(data))
            print("----")
        
```

```python
report_missingness_percentages(x_train)
```

```python
report_missingness_percentages(x_valid)
```

```python
report_missingness_percentages(x_test)
```

## Q2.
### Median Imputation

```python
median = x_train.median()
median
```

```python
from sklearn.impute import SimpleImputer
x_train_med = x_train.copy()
x_valid_med = x_valid.copy()
x_test_med = x_test.copy()
imp_med = SimpleImputer(missing_values=np.nan, strategy='median')
imp_med.fit(x_train_med) ## traine göre fit edildi
x_train_med = imp_med.transform(x_train_med)
x_valid_med = imp_med.transform(x_valid_med)
x_test_med = imp_med.transform(x_test_med)
x_train_med= pd.DataFrame(x_train_med, columns=x.columns)
x_valid_med= pd.DataFrame(x_valid_med, columns=x.columns)
x_test_med= pd.DataFrame(x_test_med, columns=x.columns)
x_test_med.head()
```

```python
x_test.head()  ## NAN values imputed by median values of training set >> look at X6,X9 and x13
```

### Mean Imputation

```python
x_train_mean = x_train.copy()
x_valid_mean = x_valid.copy()
x_test_mean = x_test.copy()
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(x_train_mean) ## traine göre fit edildi
x_train_mean = imp_mean.transform(x_train_mean)
x_valid_mean = imp_mean.transform(x_valid_mean)
x_test_mean = imp_mean.transform(x_test_mean)
x_train_mean= pd.DataFrame(x_train_mean, columns=x.columns)
x_valid_mean= pd.DataFrame(x_valid_mean, columns=x.columns)
x_test_mean= pd.DataFrame(x_test_mean, columns=x.columns)
x_test_mean.head()
```

```python
x_train.mean()  ## NAN values imputed by mean values of training set >> look at X6,X9 and x13
```
